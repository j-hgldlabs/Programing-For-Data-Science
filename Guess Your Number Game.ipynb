```python
import random

#this section picks the random secret nnumber
correct_number = random.randint(1, 5)

#I've copied and pasted the ASCII art from https://www.asciiart.eu/miscellaneous/awards & https://www.asciiart.eu/miscellaneous/dice
ascci_correct = r"""
  _______
 |       |
(| Winner|)
 |  #1   |
  \     /
   `---'
   _|_|_
* Well done! *
"""
ascci_wrong = r"""
  ____
 /\' .\    _____
/: \___\  / .  /\
\' / . / /____/..\
 \/___/  \'  '\  /
          \'__'\/
* No dice, you've lost *
"""

print("Welcome to my guess game.")
print("You can type 'I\'m done\' to exit at whenever you'd like.")

guess = ""

while True:
    guess = input(" What's your best guess??: ")

    if guess == "I'm done":
        print("No worries, thanks for playing.")
        break

    guess = int(guess)
    
    if guess == correct_number:
        print(ascci_correct)
        break
    else: 
        print(ascci_wrong)

        if guess > correct_number:
            print("Yikes, you've guessed to high.\n")
        else:
            print("Hmm - that's to low. Keep trying.\n")
```
