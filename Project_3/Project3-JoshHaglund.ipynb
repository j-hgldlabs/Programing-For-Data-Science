```python
import matplotlib.pyplot as plt
from collections import defaultdict
import string
```

```python
book = "TheKingInYellow.txt"

with open(book, 'r', encoding='utf-8') as f:
    text = f.read().lower()
```

```python
start = text.find("*** START ")
end = text.find("*** END OF THE PROJECT GUTENBERG EBOOK THE KING IN YELLOW *** ")
if start != -1 and end != -1:
    text = text[start:end]
```

```python
text = text.lower()
for p in string.punctuation:
    text = text.replace(p, "")

words = text.split()
```

```python
total_words = len(words)
unique_words = len(set(words))
print(f"This novel contains {total_words:,} total words.")
print(f"This novel contains {unique_words:,} unique words.")
```

```python
word_counts = defaultdict(int)
for word in words:
    word_counts[word] += 1

sorted_word_counts = sorted(word_counts.items(), key=lambda x: x[1], reverse=True)
```

```python
cutoff = 80  # Gives ~30 words for The King in Yellow
top_words = [(w, c) for w, c in sorted_word_counts if c >= cutoff]

print("Most Frequent Words".center(35))
print("-" * 35)
print(f"{'Word':<20}{'Count'}")
print("-" * 35)

for w, c in top_words:
    print(f"{w:<20}{c}")
```

```python
# use the existing top_words computed earlier (fix: topwords -> top_words)
word_plot = [w for w, c in top_words]
count_plot = [c for w, c in top_words]

plt.figure(figsize=(12, 6))
plt.bar(word_plot, count_plot)
plt.xlabel("Words")
plt.ylabel("Counts")
plt.xticks(rotation=75)
plt.title("Most Frequent Words in 'The King in Yellow'")
plt.tight_layout()
plt.show()
```
